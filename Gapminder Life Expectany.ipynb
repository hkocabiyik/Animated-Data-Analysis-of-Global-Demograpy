```python
import pandas as pd
df_life = pd.read_excel("data/gapminder_lifeexpectancy.xlsx")
df_life
```

```python
df_life['Life expectancy'].nunique()
df_life=df_life.rename(columns={"Life expectancy": "country"})
```

```python
import pandas as pd
df_con = pd.read_csv("data/continents.csv", sep=";")
df_con
```

```python
df_con["country"]
```

```python
import pandas as pd
final_df =pd.merge(left=df_life, right=df_con, on="country", how='inner')
final_df
```

```python
final_df.shape
```

```python
final_df_melt = pd.melt(final_df,
              id_vars=final_df.columns[[0,218]],
              value_vars=final_df.columns[1:217], 
              var_name= 'year', 
              value_name='Life_exp')
final_df_melt
```

```python
final_df_melt.dropna(inplace=True)
final_df_melt
```

```python
final_df_melt["year"].nunique()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(12,5))
ax = sns.boxplot(x="Life_exp", y="continent",orient="h", data=final_df_melt)
plt.gcf().set_facecolor('white')
```

```python
final_group= final_df_melt.groupby(['continent', 'year'])['Life_exp'].mean().to_frame()
final_group
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,5))
ax=sns.lineplot(data=final_df_melt, x="year", y="Life_exp", hue="continent")
plt.gcf().set_facecolor('white')
plt.show()
```
