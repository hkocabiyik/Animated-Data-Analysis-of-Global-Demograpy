```python
# Step 1: Load the life expectancy data from the gapminder_lifeexpectancy.xlsx file into a pandas DataFrame.
import pandas as pd

lifeexp = pd.read_excel("data/gapminder_lifeexpectancy.xlsx")
lifeexp
```

```python
#Step 2: Select a few columns with years. Note that the columns read from Excel are integer values.
lifeexp = lifeexp[[1950, 1975, 2000, 2015]]
lifeexp
```

```python
#Step 3: Draw a histogram for 2015 with the default settings.
import matplotlib.pyplot as plt

lifeexp[2015].hist()
plt.savefig('histo.png')
```

```python
#Step 4: Try a few values for the number of bins
lifeexp[2015].hist(bins=5)
lifeexp[2015].hist(bins=10)
lifeexp[2015].hist(bins=20)
```

```python
#Step 5: Modify the plot using some matplotlib functions.
plt.hist(lifeexp[2015])
plt.title('Life Expectancy& Number of Countries ')
plt.xlabel('Life Expectany')
plt.ylabel('Number of Countries')
plt.show()
```

```python
lifeexp[2015].value_counts()
```

```python
#Step 6: Set the min/max x and y values with plt.axis.
import numpy as np
lifeexp[2015].hist(alpha=0.3, histtype='bar')
plt.title('Life Expectancy& Number of Countries ')
plt.xlabel('Life Expectany')
plt.ylabel('Number of Countries')
plt.axis([lifeexp[2015].min(), lifeexp[2015].max(),0,50])
```

```python
#Step 8: Set the resolution of the image that is written.
plt.savefig('histo2.png', dpi=150)
```
