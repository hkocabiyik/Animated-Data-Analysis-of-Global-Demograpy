```python
import pandas as pd

df_life = pd.read_excel('data/gapminder_lifeexpectancy.xlsx', index_col=0)
df_life
```

```python
df_fert.columns = df_fert.columns.astype(int)
df_fert.head()
```

# Step 5

```python
df_fert.index.name = 'country'
df_fert.head()
```

# Step 6

```python
df_fert = df_fert.reset_index()
df_fert.head()
```

```python
df_fert = df_fert.melt(id_vars='country', var_name='year', value_name='fertility_rate')
df_fert.dropna(inplace=True)
df_fert
```

# Step 7

```python
df_life.index.name = 'country'
df_life.head()
df_life.columns = df_life.columns.astype(int)
```

```python
df_life = df_life.reset_index()
df_life.head()
```

```python
df_life = df_life.melt(id_vars='country', var_name='year', value_name='life_exp')
df_life.dropna(inplace=True)
df_life
```

# Step 8

```python
import pandas as pd

df_pop = pd.read_excel('data/gapminder_population.xlsx', index_col=0)
df_pop
```

```python
df_pop.index.name = 'country'
df_pop.columns = df_pop.columns.astype(int)
df_pop = df_pop.reset_index()
df_pop = df_pop.melt(id_vars='country', var_name='year', value_name='population')
df_pop.dropna(inplace=True)
df_pop
```

# Step 9

```python
final_df =pd.merge(left=df_fert, right=df_pop, on=["country","year"], how='left')
final_df
```

```python
final_df_2=pd.merge(left=final_df, right=df_life, on=["country","year"], how='left')
final_df_2
```

```python
final_df_2.isna().sum()
```

```python
final_df_2.head()
```

```python
df_subset= final_df_2.loc[final_df_2['year'] == 2000]
df_subset.reset_index(drop=True)
```

```python
len(df_subset)
```

# Step 10

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,7))
ax=sns.scatterplot(x='life_exp', y='fertility_rate',data=df_subset, alpha=0.6)
ax.axis((35, 90, 0, 10))
ax.axes.set_title("Year 2000",fontsize=35, color="b")
ax.axhline(y=df_subset["fertility_rate"].mean(), color='r', linestyle='-')
ax.axvline(x=df_subset["life_exp"].mean(), color='r', linestyle='-')
ax.axes.set_xlabel("Life expectency",fontsize=20, color="r")
ax.axes.set_ylabel("Fertility rate",fontsize=20, color="r")
# add annotations one by one with a loop
for line in range(len(df_subset)):
     plt.text(df_subset["life_exp"].iloc[line]+0.2, df_subset["fertility_rate"].iloc[line], df_subset["country"].iloc[line], horizontalalignment='left', size='small', color='g')
plt.show()
```

# Step 11 & Step 12

```python
for current_year in range(1960, 2016):
    df_subset= final_df_2.loc[final_df_2['year'] == current_year]
    plt.figure(figsize=(12,7))
    ax=sns.scatterplot(x='life_exp', y='fertility_rate', hue="country", size="population",sizes=(50, 2000), data=df_subset,legend=False)
    figname="lifeexp_"+str(current_year)+".png"
    ax.axis((35, 90, 0, 10))
    ax.axes.set_title("Year "+str(current_year),fontsize=30)
    ax.axhline(y=df_subset["fertility_rate"].mean(), color='r', linestyle='-')
    ax.axvline(x=df_subset["life_exp"].mean(), color='r', linestyle='-')
    ax.axes.set_xlabel("Life expectency",fontsize=20, color="b")
    ax.axes.set_ylabel("Fertility rate",fontsize=20, color="r")
    for line in range(len(df_subset)):
         plt.text(df_subset["life_exp"].iloc[line]+0.2, df_subset["fertility_rate"].iloc[line], df_subset["country"].iloc[line], horizontalalignment='left', size='small', color='g')
    plt.savefig(figname, dpi=300)
    plt.close()
    
```
